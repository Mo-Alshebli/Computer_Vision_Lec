```python
# How to build  Trackbar
import numpy as np

imag=np.zeros((500,512,3),np.uint8)
# imag=cv2.imread("4.png")
cv2.namedWindow("image")
def Empty(x):
    print(x)
cv2.createTrackbar('B','image',0,255,Empty)
cv2.createTrackbar('G','image',0,255,Empty)
cv2.createTrackbar('R','image',0,255,Empty)
Edit="0 / disenable: 1/ enable "
E=cv2.createTrackbar(Edit,'image',0,1,Empty)

while(True):
    cv2.imshow("image",imag)
    k=cv2.waitKey(1)
    if k==27:
        break

    B=cv2.getTrackbarPos('B','image')
    G=cv2.getTrackbarPos('G','image')
    R=cv2.getTrackbarPos('R','image')
    E=cv2.getTrackbarPos(Edit,'image')
    font=cv2.FONT_HERSHEY_SCRIPT_COMPLEX

    if E==0:
        imag[:]=0
    else:
        
        imag[:]=[B,G,R]
        cv2.putText(imag,f"{B},{G},{R}",(200,200),font,1,(200,200,0),1)



cv2.destroyAllWindows()
```

```python
# object Detection using HSV Color
#The HSV or Hue, Saturation and Value of a given object is the color space associated with the object
# in OpenCV where Hue represents the color,
# __Saturation_ represents the greyness and _Value_ represents 
# the brightness and it is used to solve the problems related to
# computer vision because of its better performance when compared to RGB or Red, 
# Blue and Green color space and the Hue range in HSV is [0,179], 
# the Saturation range in HSV is [0,255] and the Value range in HSV is [0,255] and to perform object detection, 
# finding the range of HSV is necessary.
```

```python
while(True):
    frame=cv2.imread("Trakpar.png")
    hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    
    lower_b=np.array([110,50,50])
    upper_b=np.array([130,255,255])
    
    
    mask=cv2.inRange(hsv,lower_b,upper_b)
    
    results=cv2.bitwise_and(frame,frame,mask=mask)
    cv2.imshow("Detected the Blue ",frame)
    cv2.imshow("mask",mask)
    cv2.imshow("results ",results)
    
    k=cv2.waitKey(1)
    if k==27:
        break
        

cv2.destroyAllWindows()
```

```python
# Adaptive Thresholding 
import cv2
import numpy as np
img=cv2.imread("sudoku .png",0)
# img=cv2.imread("4.png")

ret,thre_0=cv2.threshold(img,55,255,cv2.THRESH_BINARY)
thre_1=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2);
thre_2=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,111,2);
cv2.imshow("original",img)
cv2.imshow("thre_0",thre_0)
cv2.imshow("thre_1",thre_1)
cv2.imshow("thre_2",thre_2)

  
cv2.waitKey(0)
cv2.destroyAllWindows()
```
