```python
import cv2
import numpy as np 
import pandas as bd
from matplotlib import pyplot as plt
```

```python
img=cv2.imread("3.png")
cv2.imshow('image',img)
k=cv2.waitKey(0)
if k==ord('q'):
    cv2.destroyAllWindows()
else:
    print(k)
cv2.imwrite("New_Photo.png",img)
```

```python
cap=cv2.VideoCapture(0)
while(True):
    
    ret,frame=cap.read()
    print(ret)
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow("Imag",gray)
    if cv2.waitKey(1)==ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```

```python
cap=cv2.VideoCapture(0)
i=0
while(True):
    
    ret,frame=cap.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow("Imag",gray)
    print(i)
    cv2.imwrite(f"image/New_Photo{i}.png",frame)       
    i=i+1
    if cv2.waitKey(1)==ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```

```python
imag=np.zeros((512,512,3),np.uint8)
cv2.line(imag,(0,255),(511,255),(44,222,22),5)
# cv2.line(imag,(0,511),(511,0),(44,222,22),50)
# cv2.line(imag,(255,0),(255,511),(244,222,222),50)
# cv2.line(imag,(0,0),(511,511),(44,222,22),50)
cv2.imshow("ii",imag)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# imag=np.zeros((512,512,3),np.uint8)
imag=np.zeros((512,512,3),np.uint8)
cv2.rectangle(imag,(0,0),(100,100),(22,222,2),3)
cv2.rectangle(imag,(10,110),(400,400),(22,222,2),-1)

cv2.imshow("ii",imag)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
imag=np.zeros((512,512,3),np.uint8)
cv2.circle(imag,(255,255),90,(22,155,111),5)

cv2.imshow("ii",imag)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
imag=np.zeros((512,512,3),np.uint8)
g=cv2.FONT_HERSHEY_PLAIN

cv2.putText(imag,"Mohammed",(100,200),g,2,(255,255,255),5)
cv2.putText(imag,"Alshebli",(100,300),g,2,(255,255,255),5)

cv2.imshow("ii",imag)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
